```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
archivo_csv = '/content/kc_house_data.csv'  # La ruta de tu archivo CSV
datos = pd.read_csv(archivo_csv)
```

```python
# Extraer dos columnas de datos
columna1 = 'bathrooms'  # Precio
columna2 = 'price'  #  Ã¡rea habitable de una casa, expresada en pies cuadrados
datos_a_usar = datos[[columna1, columna2]]
print(datos_a_usar)
```

```python
# Aplicar el algoritmo K-Means
numero_clusters = 9  # Puedes ajustar este valor
kmeans = KMeans(n_clusters=numero_clusters)
kmeans.fit(datos_a_usar)
etiquetas = kmeans.predict(datos_a_usar)
centroides = kmeans.cluster_centers_
```

```python
# Visualizar los resultados
plt.scatter(datos_a_usar[columna1], datos_a_usar[columna2], c=etiquetas, cmap='viridis')
plt.scatter(centroides[:, 0], centroides[:, 1], marker='X', s=200, c='red')
plt.xlabel(columna1)
plt.ylabel(columna2)
plt.title('Resultados del algoritmo K-Means')
plt.show()
```
